```python
import pandas as pd
titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
```

```python
titanic.groupby('sex').sum()
```

```python
# 1. Pie Chart - Representing the male/female proportion

import matplotlib.pyplot as plt
 
# Data to plot
labels = titanic["sex"].unique()
gpbysex = titanic.groupby(['sex'])['pclass'].sum()
print(gpbysex)
colors = ['gold', 'pink']
 
# Plotting the Pie Chart 
plt.pie(gpbysex, autopct='%1.1f%%', colors=colors, shadow=True, labels = ['Male','Female'])
plt.show()
```

```python
# 2. Scatterplot with the Fare paid and the Age, differ the plot color by gender

gpbyfpa = titanic.groupby(['fare'])['age'].sum()
scatter_plot = plt.scatter(titanic['age'], titanic['fare'], s=100, alpha=0.5, c=titanic['fare'])
plt.show()
```
